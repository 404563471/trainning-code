## 表的横向拼接

### 连接表的类型

**一对一**

```python
import pandas as pd
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =0)
df1
```

```python
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =1)
df2
```

```python
#一对一连接
pd.merge(df1,df2)
```

```python
df3 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =2)
df3
```

**多对一**

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =2)
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =3)
pd.merge(df1,df2,on = "学号")
```

**多对多**

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =4)
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =3)
pd.merge(df1,df2)
```

### 连接键的类型

**默认以公共列作为连接键**

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =0)
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =1)
df1
df2
pd.merge(df1,df2)
```

**用on来指定连接键**

```python
pd.merge(df1,df2,on="学号")
```

```python
#连接键有多个
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =0)
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =5)
pd.merge(df1,df2,on=["姓名","学号"])
```

**分别指定左右连接键**  
当两个表面有公共列时，实际值一样，但列名不同。用left_on和right_on来指明连接的列名

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =6)
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =1)
pd.merge(df1,df2,left_on = "编号",right_on = "学号")
```

**把索引列当作连接键**  
索引列不算真正的列，当公共列是索引列时，就要把索引列单做连接键，使用left_index、right_index来设置索引

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =7)
df1.set_index("编号")
```

```python
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =1)
df2.set_index("学号")
```

```python
#当左右表的连接键均为索引
pd.merge(df1.set_index("编号"),df2.set_index("学号"),left_index = True,right_index = True)
```

```python
df2 = pd.read_excel(r"../Data/chapter11.xlsx",sheet_name =1)
df2
```

```python
#索引列和普通索引列混用
pd.merge(df1.set_index("编号"),df2,left_index = True,right_on = "学号")
```

### 连接方式  
用参数how来指明具体的连接方式,如果不指定默认是内连接

**内连接(inner)**

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =0)
df1
```

```python
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =8)
df2
```

```python
pd.merge(df1,df2,on="学号",how="inner")
```

**左连接**

```python
pd.merge(df1,df2,on="学号",how="left")
```

**右连接**

```python
pd.merge(df1,df2,on="学号",how="right")
```

**外连接**  
就是两个表的并集

```python
pd.merge(df1,df2,on="学号",how="outer")
```

**重复列名处理**  
pd.merge()方法默认会给这些重复的列名加后缀\_x,\_y

```python
pd.merge(df1,df2,on="学号",how="inner")
```

**自定义重复列名**  
用suffixes参数

```python
pd.merge(df1,df2,on="学号",how="inner",suffixes=["_L","_R"])
```

## 表的纵向拼接

**普通合并**

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =9)
df1
```

```python
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =10)
df2
```

```python
pd.concat([df1,df2])
```

**索引设置**

```python
pd.concat([df1.set_index("编号"),df2.set_index("编号")],ignore_index = True)
```

**重叠数据合并**

```python
df1 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =11)
df1
```

```python
df2 = pd.read_excel(r"../Data/Chapter11.xlsx",sheet_name =10)
df2
```

```python
pd.concat([df1.set_index("编号"),df2.set_index("编号")],ignore_index = True)
```

```python
#删除重复值
pd.concat([df1.set_index("编号"),df2.set_index("编号")],ignore_index = True).drop_duplicates()
```
